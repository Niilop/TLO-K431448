```python
#Luodaan datasetti, joka sisältää hinnan, sekä dataa arvosteluista
FILE = 'http://data.insideairbnb.com/germany/be/berlin/2020-03-17/data/listings.csv.gz'

cols = ['price','reviews_per_month','number_of_reviews','review_scores_rating']

df = pd.read_csv(FILE, usecols=cols, compression='infer')
df.head(5)
```

```python
#Tarkastetaan muuttujien tyypit
df.dtypes
```

```python
#tarkastellaan dataa, löytyykö sieltä virheellisiä arvoja jotka voisivat olla haitallisia
df.isnull().any()
```

```python
#review_scores_rating ja reviews_per_month sisältää null arvoja
df = df.fillna(method='ffill')
df.isnull().any()
```

```python
#muutetaan hinta float tyyppiseksi ja poistetaan niistä nolla arvot
df["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)
df = df[df.price != 0]
df.price.head(5)
```

### Koneoppiminen rooli ja sen periaatteet
Koneoppiminen on realistisesti heikkoa ja kapeaa. Usein koneoppimisohjelmalla on yksi tarkoin määritelty tehtävä, johon käytetään siihen tehtävään soveltuvaa dataa. Koneoppimisella voidaan toteuttaa erittäin rajoitetusti muutamia "älykkäitä" toimintoja. Koneoppiminen perustuu syötedatasta muodostettuihin päätelmiin. Esimerkiksi lineaarisessa regressiossa luodaan syötedatan avulla malli, jolla voidaan tehdä ennustuksia.

Tekoälyä käytetään melko laajasti yhteiskunnassa esimerkiksi terveydenhoidossa, teollisuudessa ja eri palveluiden tunnistus toiminnoissa (liike, ääni, teksti ja kuvat). Tekoäly on tällä hetkellä erittäin trendikästä ja yritysten suosiossa, mutta todellisuudessa sen järjestäminen voi olla erittäin resurssi raskasta ja siitä on vaikeaa tehdä kannattavaa liiketoimintamallia.

Koneoppiminen jaetaan eri osa-alueisiin joita ovat ohjattu oppiminen, ohjaamaton oppiminen, puoliohjattu oppiminen, vahvistusoppiminen, sekä syväoppiminen. Yleisimpinä ovat luennolla käsitelty ohjattu oppiminen, sekä ohjaamaton oppiminen.


### Ohjattu oppiminen
Ohjattu oppiminen perustuu syötedataan, eli piirteisiin. Mallille syötetään sille tarkoin määriteltyä dataa, jolloin kone pystyy tekemään niistä jotakin johtopäätelmiä. Ohjatussa oppimisessa ei suinkaan pyritä korvaamaan ihmistä, vaan halutaan opettaa kone tekemään rutiinitoimintoja. Raakadatasta erotettavat piirteet jaetaan ennustettaviin ja ennustaviin muuttujiin jotka sijoitetaan piirrematriisiin. Piirteet voivat olla binäärisiä, kategorioita tai jatkuvia. Matriisi syötetään koneoppimisalgoritmiin jolloin lopputulokseksi tulee johtopäätöksiä siitä, miten selittävät piirteet X vaikuttavat selitettävään piirteeseen Y.

Ohjattua oppimista on esimerkiksi:
- lineaarinen regressio
- letter recognition
- speech to text

### Regression vaiheet
- Erotellaan datasta piirteitä ja jalostetaan niitä sopiviksi.
- Piirteistä tehdään johtopäätöksiä: mitkä piirteet voisivat liittyä toisiinsa?
- Määritellään selitettävä muuttuja Y ja selittävä muuttuja X ja tehdään näiden perusteella mallille nollahypoteesi. Monimuuttuja regressiossa määritellään useampi X muuttuja.
- Analysoidaan lopputuloksia, tavoitteena saada malli joka selittää Y:n vaihtelua mahdollisimman hyvin. Selittäviä muuttujia voidaan karsia esimerkiksi niiden P-arvojen mukaan.
- Todetaan analyysin mukaan onko malli tilastollisesti merkitsevä ja hylätäänkö nollahypoteesi.
- Mallilla tehdään ennustuksia uusien syötteiden perusteella.

### Ohjaamaton oppiminen
Ohjaamaton oppiminen perustuu säännönmukaisuuksien etsimiseen sellaisessa ympäristössä, missä ratkaisuja ei tunneta etukäteen.

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

#Määritellää selitettävä Y ja selittävä muuttujat X
#Otetaan selittäväksi muuttujaksi tällä kertaa vain kuukautiset arvostelut
Y = df[['price']]
X = df[['reviews_per_month']]

#Luodaan opetusdata train_test_split menetelmällä
x_train,x_test,y_train,y_test = train_test_split(X,Y)

#Luodaan lineaarinen malli ja opetetaan sitä datalla
LR = linear_model.LinearRegression()
LR.fit(x_train.values.reshape(-1,1),y_train.values)

#Tehdään ennustus ja katsotaan miltä tulokset näyttävät kuvaajalla
prediction = LR.predict(x_test.values.reshape(-1,1))
plt.plot(x_test, prediction, label='Linear Regression', color = 'b')
plt.scatter(x_test, y_test, label='Testing data', color='g')
plt.legend()
plt.show()
```

```python
teksti = ['Tämä On Esimerkki Jota Jalostetaan',
         'Tämä Lause Sisältää ErikoisMerkkejä ¤¤%%&&$$££@@',
         'ammatti-ihminen ala-aste linja-auto tasa-arvo-ohjelma']
df = pd.DataFrame(teksti)
df
```

```python
import re
#Luodaan tyhjä lista lauseille
prosessoidut_lauseet = []

#Käydään kaikki lauseet yksitellen läpi ja splitataan lauseet omiksi sanoiksi jota prosessoidaan
for lause in teksti:
    
    #Luodaan tyhjä lista sanoille
    prosessoidut_sanat = []
    sanat = lause.split(" ")
    for sana in sanat:
        if len(sana) > 2:
            sana = re.sub("[^A-Za-z0-9ÄäÖö-]+", "", sana)
            sana = re.sub("-", " ", sana)
            sana = sana.strip().lower()
            prosessoidut_sanat.append(sana)
            
    #Lisätään kaikki sanat takaisin lauseeksi ja lisätään ne listaan       
    prosessoidut_lauseet.append(" ".join(prosessoidut_sanat)) 
            
prosessoidut_lauseet
```

```python
#Luodaan sama dataframe kuin esimerkissä
df = pd.read_csv("movies_metadata.csv", low_memory=False)
dfNum = df[['popularity','runtime','vote_average','vote_count','title','overview']]

#Poistetaan null arvot
dfNum=dfNum.dropna()

#Tarkastellaan tyyppejä
dfNum.info()
```

```python
from sklearn import preprocessing

#budget ja popularity ovat object tyyppisiä, joten ne täytyy muuttaa floatiksi
#Dataframessa on kuitenkin muita string tyyppisiä muuttujia, joten ne voidaan ottaa talteen omaan dataframeen

#uusi dataframe
dfTitleOverview=dfNum[['title', 'overview']]
#poistetaan title ja overwiev
dfNum =dfNum.drop(['title','overview'], axis=1)

#muutetaan kaikki muuttujat floatiksi
dfNum = dfNum.astype(float)

#normalisoidaan muuttujat

minmax= preprocessing.MinMaxScaler().fit_transform(dfNum)
dfNumNorm = pd.DataFrame(minmax, index=dfNum.index, columns=dfNum.columns)
```

```python
from sklearn.cluster import KMeans

#Tehdään kmeans malli jossa on 3 klusteria
kmeans = KMeans(n_clusters=3)
kmeans.fit(dfNumNorm)

#lisätään uusi sarake joka kertoo sen, mihin klusteriin kuuluu
dfNum['cluster'] = kmeans.labels_

dfNum.head(5)
```

### 5. Kokoa noin kymmenen koodirivin demo viikon aikana oppimastasi.

Muutetaan aikaisempaa dataframea kaksiulotteiseksi, jotta siitä voidaan tehdä visualisointia.
Katsotaan miltä kuvaaja näyttää 10 klusterin kanssa ja 3 klusterin kanssa.

```python
df =dfNum.drop(['vote_average','vote_count', 'cluster'], axis=1)
df
```

```python
def clustering(n):
    
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(df)

    #Tehdään ennustus kmeansilla
    label = kmeans.predict(df)

    import numpy as np
    import matplotlib.pyplot as plt

    u_labels = np.unique(label)

    for i in u_labels:
        label_i = df[label == i]

        plt.scatter(label_i['popularity'] , label_i['runtime'])
    plt.legend()
    plt.show()
```

```python
clustering(10)
```

```python
clustering(3)
```
